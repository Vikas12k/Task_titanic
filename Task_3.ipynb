```python
file_path = os.path.join(path, 'Housing.csv')
df = pd.read_csv(file_path)

# Handling missing values for available columns
for col in ['bedrooms', 'bathrooms']: # Corrected columns to impute
    median_val = df[col].median()
    df[col].fillna(median_val, inplace=True)

# Convert categorical features to dummy variables for available columns
df_processed = pd.get_dummies(df, columns=['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea', 'furnishingstatus'], drop_first=True) # Corrected columns for one-hot encoding

display(df_processed.head())
display(df_processed.isnull().sum())
```

```python
for col in ['bedrooms', 'bathrooms']:
    median_val = df[col].median()
    df[col].fillna(median_val, inplace=True)

df_processed = pd.get_dummies(df, columns=['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea', 'furnishingstatus'], drop_first=True)

display(df_processed.head())
display(df_processed.isnull().sum())
```

```python
from sklearn.model_selection import train_test_split

y = df_processed['price']

X_simple = df_processed[['area']]

X_multi = df_processed.drop('price', axis=1)

X_simple_train, X_simple_test, y_simple_train, y_simple_test = train_test_split(
    X_simple, y, test_size=0.2, random_state=42)

X_multi_train, X_multi_test, y_multi_train, y_multi_test = train_test_split(
    X_multi, y, test_size=0.2, random_state=42)

display(X_simple_train.head())
display(X_simple_test.head())
display(y_simple_train.head())
display(y_simple_test.head())
display(X_multi_train.head())
display(X_multi_test.head())
display(y_multi_train.head())
display(y_multi_test.head())
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

simple_model = LinearRegression()
simple_model.fit(X_simple_train, y_simple_train)

y_simple_pred = simple_model.predict(X_simple_test)

mae_simple = mean_absolute_error(y_simple_test, y_simple_pred)
mse_simple = mean_squared_error(y_simple_test, y_simple_pred)
r2_simple = r2_score(y_simple_test, y_simple_pred)

print(f"Simple Linear Regression MAE: {mae_simple}")
print(f"Simple Linear Regression MSE: {mse_simple}")
print(f"Simple Linear Regression R2: {r2_simple}")
```

```python
multiple_model = LinearRegression()
multiple_model.fit(X_multi_train, y_multi_train)

y_multi_pred = multiple_model.predict(X_multi_test)

mae_multi = mean_absolute_error(y_multi_test, y_multi_pred)
mse_multi = mean_squared_error(y_multi_test, y_multi_pred)
r2_multi = r2_score(y_multi_test, y_multi_pred)

print(f"Multiple Linear Regression MAE: {mae_multi}")
print(f"Multiple Linear Regression MSE: {mse_multi}")
print(f"Multiple Linear Regression R2: {r2_multi}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10, 6))
plt.scatter(X_simple_test, y_simple_test, color='blue', label='Actual Prices')
plt.plot(X_simple_test, y_simple_pred, color='red', linewidth=2, label='Regression Line')
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Simple Linear Regression: Area vs. Price")
plt.legend()
plt.show()
```

```python
print("Simple Linear Regression Coefficient:")
simple_coef = simple_model.coef_[0]
print(f"The coefficient for 'Area' is: {simple_coef}")
print("Interpretation: For every one unit increase in 'Area', the predicted 'Price' increases by approximately", simple_coef, ".")
print("\nMultiple Linear Regression Coefficients:")
multi_coefs = multiple_model.coef_
feature_names = X_multi_train.columns

for feature, coef in zip(feature_names, multi_coefs):
    print(f"Feature: {feature}, Coefficient: {coef}")

print("\nInterpretation: Each coefficient represents the estimated change in 'Price' for a one-unit increase in the corresponding feature, holding all other features constant.")
```

```python
print("--- Model Performance Comparison ---")
print(f"Simple Linear Regression MAE: {mae_simple:.2f}")
print(f"Multiple Linear Regression MAE: {mae_multi:.2f}")
print("\nSimple Linear Regression MSE: ", f"{mse_simple:.2f}")
print("Multiple Linear Regression MSE: ", f"{mse_multi:.2f}")
print("\nSimple Linear Regression R2: ", f"{r2_simple:.4f}")
print("Multiple Linear Regression R2: ", f"{r2_multi:.4f}")

print("\n--- Performance Summary ---")
if mae_multi < mae_simple:
    print("The Multiple Linear Regression model has a lower MAE, indicating better accuracy in predictions.")
else:
    print("The Simple Linear Regression model has a lower MAE, indicating better accuracy in predictions.")

if mse_multi < mse_simple:
    print("The Multiple Linear Regression model has a lower MSE, indicating smaller average squared errors.")
else:
    print("The Simple Linear Regression model has a lower MSE, indicating smaller average squared errors.")

if r2_multi > r2_simple:
    print("The Multiple Linear Regression model has a higher R², indicating that it explains more of the variance in the target variable.")
else:
    print("The Simple Linear Regression model has a higher R², indicating that it explains more of the variance in the target variable.")

print("\nOverall, the Multiple Linear Regression model performed better than the Simple Linear Regression model based on all three metrics (MAE, MSE, and R²). This is likely because it incorporates more relevant features, allowing it to capture a more complex relationship between the features and the house prices.")
```
